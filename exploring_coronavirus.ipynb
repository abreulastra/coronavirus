```python
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use('classic')
import seaborn as sns
#sns.set(style="whitegrid")
#sns.set(rc={'figure.figsize':(20,8.27)})

plt.rcParams.update({'font.size': 16});
```

```python
# These are the urls from the github repossitory updated every day
url_cases = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
url_recoveries = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'
url_deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv'

# Then we extract the urls to make pandas dataframes
cases = pd.read_csv(url_cases).drop(columns = ['Province/State', 'Lat', 'Long'])
recoveries = pd.read_csv(url_recoveries).drop(columns = ['Province/State', 'Lat', 'Long'])

dates = cases.columns[1:]
cases = cases.groupby('Country/Region')[dates].sum()
recoveries = recoveries.groupby('Country/Region')[dates].sum()

deaths = pd.read_csv(url_deaths).drop(columns = ['Province/State', 'Lat', 'Long'])
dates = cases.columns[1:]
deaths = deaths.groupby('Country/Region')[dates].sum()

cases = cases.groupby('Country/Region')[dates].sum()
recoveries = recoveries.groupby('Country/Region')[dates].sum()
deaths = deaths.groupby('Country/Region')[dates].sum()
```

```python
def country(coun):
    country = pd.DataFrame()
    country['Cumulative cases'] = cases.T[coun]
    country['Cumulative recoveries'] = recoveries.T[coun]
    country['Cumulative deaths'] = deaths.T[coun]
    
    country['Active cases'] = country['Cumulative cases'] - country['Cumulative recoveries'] - country['Cumulative deaths']
    title = 'Coronavirus cases in ' + coun
    plot = country.tail(25).plot(title = title, figsize=(20,10)).figure.savefig(coun);
```
